```python
import numpy

def p(T):
    
    return (999.85308 + 6.3269*10**(-2)*T - 8.523829*10**(-3)*T**2 + 6.943248*10**(-5)*T**3 -3.821216*10**(-7)*T**4)

def p1(T):
    
    return (999.85308 + 6.3269*10**(-2)*T - 8.523829*10**(-3)*T**2 + 6.943248*10**(-5)*T**3)

T=[17.1,17.5]

def std_dev(X):

    mean = sum(X)/len(X)

    variance = sum((x - mean) ** 2 for x in X) / (len(X) - 1)

    return variance ** 0.5

print('Desnity at 17.3 degrees' , p(17.3))
print('Desnity at 17.3 degrees1' , p1(17.3))
print('Error in density', abs(p(17.3+std_dev(T)/numpy.sqrt(2))-p(17.3)))
print('Error in density', abs(p1(17.3+std_dev(T)/numpy.sqrt(2))-p1(17.3)))
```
